```python
# Let's import the pandas, numpy libraries as pd, and np respectively. 
import pandas as pd
import numpy as np

# Load the pyplot collection of functions from matplotlib, as plt 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

```python
properties
```

```python
#looking at the arrangement of the output,I feel to rotate the data from columns into rows and viceversa
properties = properties.transpose()
```

```python
properties.columns = properties.iloc[0]
properties = properties.iloc[1:]
properties = properties.reset_index()
#properties=properties.loc[:,~properties.columns.str.contains('^Unnamed')]
```

```python
properties.rename(columns={'index': 'London_Borough', properties.columns[1]: 'EID'}, inplace=True)
properties.columns.name = '' #Now Unnamed:0 makes no sense ,so its better to drop that
properties.head()



```

```python
#Removing rows with all Nans
properties=properties.dropna()
```

```python
properties = pd.melt(properties, id_vars=['EID', 'London_Borough'], var_name='Year/Month', value_name='Housing Price')
```

```python
properties['Housing Price'] = pd.to_numeric(properties['Housing Price'])
properties.dtypes
properties.head(10)
```

```python
barkingDagenham_df = properties[properties['London_Borough'] == 'Barking & Dagenham']
barkingDagenham_df.plot(x='Year/Month', y='Housing Price')
plt.ylabel('Housing Price')
plt.title('Barking & Dagenham Price Fluctuations')
plt.show()
```

```python
def create_price_ratio(df, District) :
  df_borough = df[df['London_Borough'] == District]
  avg_1998 = df_borough[df_borough['Year/Month'].dt.year == 1998]['Housing Price'].mean()
  avg_2018 = df_borough[df_borough['Year/Month'].dt.year == 2018]['Housing Price'].mean()
  return avg_2018/avg_1998

London_Boroughs= properties['London_Borough'].unique()
results = []
for District in London_Boroughs :
  results.append((str(District), create_price_ratio(properties, District)))
sorted_results = sorted(results, key = lambda x : x[1], reverse=True)
for i in range(10) : final=print(sorted_results[i])
```
